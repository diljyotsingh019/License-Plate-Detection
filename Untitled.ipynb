```python
import easyocr
from ultralytics import YOLO
import os
import numpy as np
import cv2
```

```python
os.listdir()
```

```python
reader = easyocr.Reader(['en'])
```

```python
cap = cv2.VideoCapture("pexels_videos_2103099 (2160p).mp4")
model = YOLO("runs/detect/train/weights/best.pt")

while True:
    ret, frame = cap.read()
    if ret == False:
        break
        
    results = model.predict(frame)
    for result in results[0].boxes.xyxy:
        x1, y1, x2, y2 = result
        cv2.rectangle(frame,(int(x1), int(y1)), (int(x2), int(y2)), color = (0,255,0), thickness = 5)
        ocr = reader.readtext(frame[int(y1):int(y2), int(x1):int(x2)])
        text=""
        print(ocr)
        for i in ocr:
            if i:
                if i[2]>0.1:
                    text +=i[1]
        cv2.putText(frame, text.upper(), (int(x1), int(y1)-30), 2,1,(0,255,0), 4, cv2.LINE_AA)
    frame = cv2.resize(frame, (960,540))
    cv2.imshow("License Plate detection", frame)
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
